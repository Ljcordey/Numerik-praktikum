```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

data = np.genfromtxt('data.txt')
```

```python
T = range(data.shape[1])
print(data[0])
plt.plot(data)

plt.show()
```

This first plot allows us to have a first overview of the data and their meanings.
There are two types of data points:
- one of them follows a linear scale
- the other one follows a periodical pattern

We can assume that the linear scale is for the **time** data points and the periodical pattern is for the actual measurement we want to analyse.

```python
plt.plot(data[:,0],data[:,1])
```
